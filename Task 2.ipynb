## Introducing Scikit Learn for Creating a Decision Tree Classifier 

#### Separate Label and Features i.e. divide the data into "features" (inputs) and "label" (output).

```python
X = data.drop('species', axis=1)
y = data['species']
```

#### Train | Test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

#### importing the algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model = DecisionTreeClassifier()
```

### Using GridSearchCV for the search of best hyperparameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
max_depth = np.linspace(1,10,10,dtype =int)
max_leaf_nodes = np.linspace(1,10,10,dtype =int)

param_grid = {'max_depth': max_depth, 'max_leaf_nodes': max_leaf_nodes}
```

```python
grid_model = GridSearchCV(estimator = base_model, param_grid = param_grid,cv=5) ## GridSearchCV model
```

```python
grid_model.fit(X_train, y_train) #fit the GridSearchCV model on training data
```

```python
grid_model.best_params_ #Best Hyperparameters
```

### Creating a model by using the optimized hyperparameters given by the GridSearchCV model

```python
model = DecisionTreeClassifier(max_depth = 3, max_leaf_nodes = 4)
```

```python
model.fit(X_train, y_train) #fit the model
```

#### Prediction on testing data

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
print('Actual vs Predicted')
class_res = pd.DataFrame({'y_pred':y_pred, 'y_true': y_test})
class_res
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance']) #Important Features
```

### Evaluation

```python
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(model, X_test, y_test)
plt.show()
```

### Visualise the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(3,4), dpi=200)
plot_tree(model, feature_names= X.columns, filled=True)
plt.show()
```

### Retraining the model on full data

```python
model.fit(X, y)
```

```python
plt.figure(figsize=(3,4), dpi=200)
plot_tree(model, feature_names= X.columns, filled=True)
plt.show()
```
